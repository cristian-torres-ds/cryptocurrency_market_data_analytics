Al analizar y recomendar criptomonedas, hay varias variables clave que hay que considerar. 

Estas son algunas de las más importantes:

1. **Capitalización de mercado**: Es el valor total de una criptomoneda en circulación. Las criptomonedas con una alta capitalización de mercado suelen ser más estables y tienen más liquidez.

2. **Volumen de negociación**: El volumen diario de transacciones es importante para determinar la liquidez y la actividad en el mercado de una criptomoneda.

3. **Tecnología y Funcionalidad**: Evalúa la tecnología subyacente de la criptomoneda. ¿Qué problemas resuelve? ¿Qué innovaciones aporta? ¿Tiene una propuesta de valor única?

4. **Equipo y Desarrollo**: Investiga el equipo de desarrollo detrás de la criptomoneda. ¿Son expertos en el campo? ¿Tienen un historial de éxito en proyectos anteriores?

5. **Adopción y Uso**: ¿Qué tan ampliamente se utiliza la criptomoneda en el mundo real? ¿Es aceptada en diferentes industrias o casos de uso?

6. **Competencia y Diferenciación**: Compara la criptomoneda con otras en su categoría. ¿Tiene características únicas o ventajas sobre sus competidores?

7. **Regulación y Cumplimiento**: La regulación gubernamental puede tener un impacto significativo en el valor y la adopción de una criptomoneda. Evalúa la posición de la criptomoneda con respecto a la regulación.

8. **Comunidad y Participación**: Una comunidad activa y comprometida puede impulsar la adopción y el desarrollo continuo de una criptomoneda.

9. **Seguridad y Historial de Seguridad**: Investigar la seguridad de la criptomoneda y si ha tenido problemas de seguridad en el pasado.

10. **Perspectivas a largo plazo**: Considera las perspectivas a largo plazo de la criptomoneda. ¿Tiene un caso de uso sostenible y una visión a largo plazo?

11. **Noticias y Eventos**: Mantente al tanto de las noticias y eventos que puedan afectar el valor y la adopción de la criptomoneda.

```python
import pandas as pd
import datetime as dt
import time as t
# import plotly.graph_objects as go
# from plotly.offline import plot
from pycoingecko import CoinGeckoAPI
```

```python
#create a client
cg = CoinGeckoAPI()

#confirm connection
cg.ping()
```

```python
#get a list of coins, sort df by id
coinList = cg.get_coins_list()
coinDataFrame = pd.DataFrame.from_dict(coinList).sort_values('id'
                                      ).reset_index(drop=True)
coinDataFrame
```

```python
coins = ['bitcoin','ethereum','dopex']
```

```python
simplePriceRequest = cg.get_price(ids = coins, vs_currencies = 'usd')
print(simplePriceRequest)

simplePriceRequest
```

```python
complexPriceRequest = cg.get_price(ids = coins, 
                        vs_currencies =  'usd', 
                        include_market_cap = True,
                        include_24hr_vol = True,
                        include_24hr_change = True,
                        include_last_updated_at = True)
print(complexPriceRequest)
```

```python
assetPlatformsList = cg.get_asset_platforms()
assetPlatforms = pd.DataFrame.from_dict(assetPlatformsList
                   ).sort_values('id').reset_index(drop=True)
assetPlatforms
```

```python
coinCategoriesData = pd.DataFrame(cg.get_coins_categories(
                                                order = 'market_cap_desc'))

coinCategoriesDataFrame = pd.DataFrame.from_dict(coinCategoriesData
                                ).sort_values('market_cap').reset_index(drop=True)

coinCategoriesDataFrame
```

```python
coinCategoriesDataFrame.head(50)
```

```python
coins = coinCategoriesDataFrame.id.head(10).tolist()
coins
```

```python
#get coins market cap, rank, prices, volume, market data, etc.
#by ID using a list
coinsMktDataByIds = cg.get_coins_markets(vs_currency = 'usd',
                                        ids = coins)

coinsMktDataByIds
```

```python
#by category see cg.get_coins_categories_list(), in order
coinsMktDataByCategory = cg.get_coins_markets(vs_currency = 'usd',
                                    category = 'stablecoins',
                                    order = 'volume_desc')
coinsMktDataByCategory
```

```python
coinsMktDataByCategory = pd.DataFrame.from_dict(coinsMktDataByCategory
                                ).sort_values('market_cap').reset_index(drop=True)
coinsMktDataByCategory
```
